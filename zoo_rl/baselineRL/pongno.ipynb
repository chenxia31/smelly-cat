```python
import gymnasium as gym
import torch
import numpy as np
from sac import SAC
from replay_memory import ReplayMemory

# 创建环境
env = gym.make("Humanoid-v4")

# 获取环境相关参数
action_dim = env.action_space.shape[0]
state_dim = env.observation_space.shape[0]
hidden_dim = 256
action_range = 1.

# 设置随机数种子，确保结果可复现
seed = 12345
torch.manual_seed(seed)
np.random.seed(seed)

# 创建SAC模型
memory = ReplayMemory(1000000)
sac = SAC(state_dim, action_dim, hidden_dim, action_range)

# 训练模型
num_episodes = 1000
for i_episode in range(num_episodes):
    state, _ = env.reset()
    episode_reward = 0
    for t in range(1, 10000):  # 不要让一次回合进行太久
        action = sac.select_action(state)
        next_state, reward, done,crack,info = env.step(action)
        mask = 1 if done else 0
        memory.push(state, action, mask, next_state, reward)
        episode_reward += reward
        if len(memory) > 1000:
            sac.update_parameters(memory)
        if done:
            break
        state = next_state
    print("Episode: {}, reward: {}".format(i_episode, episode_reward))
```
